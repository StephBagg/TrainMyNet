```python
import tensorflow as tf
import numpy as np
```

```python
i = tf.placeholder(tf.float32)
with tf.name_scope('entire-board'):
    with tf.name_scope('first-operation'):
        a = tf.add(i,4.0, name='a')
        a_scalar = tf.summary.scalar('a-value', a)
        b = tf.multiply(a, 10.0, name='b')

    with tf.name_scope('second-operation'):
        c = tf.subtract(i,3.0, name='c')
        c_scalar = tf.summary.scalar('c-value', c)
        d = tf.divide(c, 9.0, name='d')

    with tf.name_scope('final-operation'):
        e = tf.add(b, d, name='e')
    
e_i = tf.summary.scalar('value_of-e',e)

merge = tf.summary.merge_all()

sess = tf.Session()
#no need to initialize variables as we only have constants.
```

```python
write = tf.summary.FileWriter('./board', sess.graph)
#write = tf.summary.FileWriter('./board')

run = True
j = 1.0
step = 0
while run==True:
    step += 1
    j += 1e-6
    e_, summary_ = sess.run([e, merge], feed_dict={i:j})
    
    if step%1000 == 0:
        print('step:', step)
        write.add_summary(summary_, step)
    
```
